## Regression exercises

Linear regression is one of the most important modelling techniques. We looked into some of the important concepts like fitting a linear model, how variation os explained by the models, error of the estimates and little bit of multiple and multivariate regression. 

In this exercise we will work with online news popularity dataset. The dataset has 61 attributes: 58 predictive attributes, 2 non-predictive, 1 response variable. We have to predict the number of shares for the news article using rest of the predictor variables. Lets start by reading the data into a datframe called news_data.

```r
news_data = read.csv("../../../datasets/OnlineNewsPopularity/OnlineNewsPopularity.csv",header=TRUE)
head(news_data)
```

The first two columns are non predictors.

**Activity 1: ** Get rid of the columns 'url', 'timedelta' from the dataframe news_data.

```r
news_data=news_data[,!names(news_data) %in% c("url","timedelta")]
head(news_data)
```

```r
summary(news_data)
```

Some of the variables are highly skewed. So what are variables which are highly correlated with the response variable.

```r
cor(news_data)>0.4
```

None of the predictors have any significant correlation with the response variable. Lets go ahead and fit the model using all the predictors. Name the model 'news_reg'

```r
news_reg=lm(shares ~., data=news_data)
summary(news_reg)
```

Well the model did ver bad. Its multiple R-squared is almost equal to 0. Lets try to build the model with predictors that are significant based on their p-values as listed below based on above model summary.

1 - n_unique_tokens, average_token_length, data_channel_is_bus

2 - n_tokens_title, n_tokens_content, num_self_hrefs, data_channel_is_lifestyle, global_subjectivity

3 - num_hrefs, data_channel_is_entertainment, kw_min_avg, kw_max_avg,kw_avg_avg, self_reference_min_shares

```r
news_reg1 = lm(shares ~ num_hrefs+ data_channel_is_entertainment+ kw_min_avg+ kw_max_avg+
               kw_avg_avg+ self_reference_min_shares,
             data=news_data)
summary(news_reg1)
```
